```python
x= data_set.iloc[:, :-1].values
y= data_set.iloc[:,  -1].values
```

```python
labelencoder_x= LabelEncoder()
x[:, 3]= labelencoder_x.fit_transform(x[:,3])
onehotencoder= OneHotEncoder()
x= onehotencoder.fit_transform(x).toarray()
```

```python
x = x[:, 1:]
```

```python
x_train_val, x_test, y_train_val, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
x_train, x_val, y_train, y_val = train_test_split(x_train_val, y_train_val, test_size=0.25, random_state=42)  # 60% training, 20% validation, 20% testing
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred= model.predict(x_test)
```

```python
print('Train Score\t: ', model.score(x_train, y_train))
print('Test Score\t: ', model.score(x_test, y_test))
```
